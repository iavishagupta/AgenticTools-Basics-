```python
import os
from google.colab import userdata

os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')
```

```python
from crewai import Agent, Crew, Task
from crewai_tools import PDFSearchTool
```

```python
tool = PDFSearchTool(pdf="/content/Fertilizer_Recommendation.pdf")
```

### Crop Names Lister

```python
point_lister = Agent(
    role="Point Lister",
    goal="You are an agent who analyzes the documents, and makes a presents of all the crop names, whose fertilizer requirement details are given in the document.",
    backstory="You are an expert document analyzer who analyzes the documents",
    verbose=True,
    allow_delegation=False,
    tools=[tool]
)
```

```python
listing = Task(
    description=f"Return all the crop names, whose fertilizer requirement details are given in the document.",
    expected_output="Output should be names of all the crop names, whose fertilizer requirement details are given in the document.",
    agent=point_lister,
    verbose=True
)
```

```python
listing_crew = Crew(
    tasks=[listing],
    agents=[point_lister],
    verbose=True
)
```

```python
crop_list = listing_crew.kickoff()
```

```python
with open("Crop_List.txt", "w+") as f:
  f.writelines(crop_list.raw)
```

```python
with open("Crop_List.txt", "r") as f:
  text = f.readlines()

text = "".join(text)
print(text)
```

### Crop Fertilizer Recommendation

```python
query = "What type of fertilizers should be used for Sunflower?"
```

```python
document_searcher = Agent(
    role="Document Searcher",
    goal="You are a document searcher, who analyzes the whole document and answers to all the questions asked by the viewer.",
    backstory="You are an expert document analyzer/searcher who searches and gives the answers to the asked questions in a concise manner.",
    verbose=True,
    allow_delegation=False,
    tools=[tool]
)
```

```python
searching = Task(
    description=f"""Answer to the queries asked by the user, analyze the document in detail, and answer to the
                    questions in a consolidated manner, not too long, not too short, but detailed enough.
                    Query:{query}""",
    expected_output="Output should be an answers in max 2 to 3 paragraphs, with all the detailes required by the asked query.",
    agent=document_searcher,
    verbose=True

)
```

```python
search_crew = Crew(
    tasks=[searching],
    agents=[document_searcher],
    verbose=True
)
```

```python
search_result = search_crew.kickoff()
```

```python
with open("Recommendation.txt", "w+") as f:
  f.writelines(search_result.raw)
```

```python
with open("Recommendation.txt", "r") as f:
  text = f.readlines()
```
